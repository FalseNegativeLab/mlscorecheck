```python
def check_accuracy(p, n, acc, eps, numerical_tolerance=1e-6):
    eps = eps + numerical_tolerance
    tptn = Interval(acc-eps, acc+eps) * (p + n)
    tptn = tptn.shrink_to_integers()
    
    if tptn.is_empty():
        return True, set()
    
    pairs = set()
    for tptn_val in range(int(tptn.lower_bound), int(tptn.upper_bound) + 1):
        for tp in range(0, min(tptn_val, p)):
            tn = tptn_val - tp
            if tn <= n:
                pairs.add((tp, tn))
    
    return len(pairs) == 0, pairs

def check_sensitivity(p, n, sens, eps, numerical_tolerance=1e-6):
    eps = eps + numerical_tolerance
    tp = Interval(sens-eps, sens+eps) * p
    tp = tp.shrink_to_integers()
    
    if tp.is_empty():
        return True, set()
    
    pairs = set()
    
    for tp_val in range(int(tp.lower_bound), int(tp.upper_bound) + 1):
        for tn in range(0, n+1):
            pairs.append(tp_val, tn)

    return len(pairs) == 0, pairs

def check_sensitivity(p, n, spec, eps, numerical_tolerance=1e-6):
    eps = eps + numerical_tolerance
    tn = Interval(spec-eps, spec+eps) * p
    tn = tp.shrink_to_integers()
    
    pairs = set()
    
    for tn_val in range(int(tn.lower_bound), int(tn.upper_bound) + 1):
        for tp in range(0, p+1):
            pairs.append(tp, tn_val)

    return len(pairs) == 0, pairs

def check_positive_predictive_value(p, n, ppv, eps, numerical_tolerance=1e-6):
    eps = eps + numerical_tolerance
    ppv_int = Interval(ppv-eps, ppv+eps)
    
    pairs = set()
    
    for tp_plus_fp in range(0, p+n):
        tp = ppv_int * tp_plus_fp
        for tp_val in range(int(tp.lower_bound), int(tp.upper_bound) + 1):
            fp = tp_plus_fp - tp_val
            tn = n - fp
            pairs.add(tp_val, tn)
        
    return len(pairs) == 0, pairs

def check_negative_predictive_value(p, n, npv, eps, numerical_tolerance=1e-6):
    eps = eps + numerical_tolerance
    npv_int = Interval(npv-eps, npv+eps)
    
    pairs = set()
    
    for tn_plus_fn in range(0, p+n):
        tn = npv_int * tn_plus_fn
        for tn_val in range(int(tn.lower_bound), int(tn.upper_bound) + 1):
            fn = tn_plus_fn - tn_val
            tp = p - fn
            pairs.add(tp, tn_val)
        
    return len(pairs) == 0, pairs

def check_f1_plus(p, n, npv, eps, numerical_tolerance=1e-6):
    eps = eps + numerical_tolerance
    npv_int = Interval(npv-eps, npv+eps)
    
    pairs = set()
    
    for tn_plus_fn in range(0, p+n):
        tn = npv_int * tn_plus_fn
        for tn_val in range(int(tn.lower_bound), int(tn.upper_bound) + 1):
            fn = tn_plus_fn - tn_val
            tp = p - fn
            pairs.add(tp, tn_val)
        
    return len(pairs) == 0, pairs

```

```python
import copy

def generate_datasets_with_all_kfolds(dataset_specification):
    if 'folds' in dataset_specification:
        raise ValueError('do not specify the "folds" key for the generation of folds')
    if ('p' in dataset_specification and 'n' not in dataset_specification)\
        or ('p' not in dataset_specification and 'p' in dataset_specification):
        raise ValueError('either specifiy both "p" and "n" or None of them')
    if 'name' in dataset_specification and ('p' in dataset_specification):
        raise ValueError('either specificy "name" or "p" and "n"')
    
    if 'name' in dataset_specification:
        p = dataset_statistics[dataset_specification["name"]]["p"]
        n = dataset_statistics[dataset_specification["name"]]["n"]
    else:
        p = dataset_specification["p"]
        n = dataset_specification["n"]
        
    kfolds = create_all_kfolds(p=p, n=n, n_folds=dataset['n_folds'])
    results = []
    for positives, negatives in zip(*kfolds):
        results.append({'folds': [{'p': p_, 'n': n_} for p_, n_ in zip(positives, negatives)]})
    
    if dataset['n_repeats'] == 1:
        return results

    all_results = copy.deepcopy(results)
    
    n_repeats = dataset['n_repeats'] - 1
    while n_repeats > 0:
        tmp = []
        for result in all_results:
            for res in results:
                result = copy.deepcopy(result)
                result['folds'].extend(res['folds'])
                tmp.append(result)
        all_results = tmp
        n_repeats -= 1
    
    return all_results
            
```

```python
def convert_to_equivalent_acc_problems(p, n, n_folds, n_repeats=1):
    n_items = (p + n) // n_folds
    n_high_fold = (p + n) % n_folds
    n_low_fold = n_folds - n_high_fold
    
    return [{'total': n_items * n_low_fold * n_repeats},
            {'total': n_items * n_high_fold * n_repeats}]
    
```

```python
def fold_variations(n_items, n_folds):
    if n_folds == 1:
        return 1, [[n_items]]
    if n_items == n_folds:
        return 1, [[1]*n_items]
    
    upper_bound = n_items - n_folds + 1
    lower_bound = int(np.ceil((n_items - n_folds + 2)/2))
    
    total = 0
    all_configurations = []
    
    for value in reversed(list(range(lower_bound, upper_bound+1))):
        count, configurations = fold_variations(n_items - value, n_folds-1)
        total += count
        for conf in configurations:
            conf.append(value)
        all_configurations.extend(configurations)
    
    return total, all_configurations
    
```

```python
def remainder_variations(remainder, n_folds):
    indices = itertools.combinations(list(range(n_folds)), remainder)
    lists = []
    for index in list(indices):
        tmp = [0] * n_folds
        for idx in index:
            tmp[idx] = 1
        lists.append(tmp)
    return lists
```

```python
def create_foldings(p, n, n_folds):
    total = p + n
    items_per_fold = total // n_folds
    remainder = total % n_folds
    
    totals = np.array(remainder_variations(remainder, n_folds)) + items_per_fold
    folds = np.array(fold_variations(p, n_folds)[1])
    
    positive_counts = []
    negative_counts = []
    
    for total_counts in totals:
        other = total_counts - folds
        other_counts = np.sum(other, axis=1)
        positive_counts.append(folds[other_counts <= n])
        negative_counts.append(other[other_counts <= n])
    
    return np.vstack(positive_counts), np.vstack(negative_counts)
```
