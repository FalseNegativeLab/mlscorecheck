```python
import numpy as np
```

```python
from mlscorecheck.machine_learning import problems_to_check, check
from mlscorecheck.core import (accuracy, sensitivity, specificity, 
                               positive_predictive_value, 
                               negative_predictive_value)
```

```python
p = 100
n = 300
tp = 80
tn = 250
fp = n - tn
fn = p - tp
k = 3
eps = 10**(-k)
acc = np.round(accuracy(tp=tp, tn=tn, total=p + n), k)
sens = np.round(sensitivity(tp=tp, p=p), k)
spec = np.round(specificity(tn=tn, n=n), k)
ppv = np.round(positive_predictive_value(tp=tp, fp=fp), k)
npv = np.round(negative_predictive_value(tn=tn, fn=fn), k)
```

```python
check(acc=acc, sens=sens, spec=spec, ppv=ppv, npv=npv, eps=eps, p=p, n=n, return_details=True)
```
