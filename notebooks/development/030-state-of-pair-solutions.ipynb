```python
import pandas as pd
pd.set_option('display.max_columns', None)
from mlscorecheck.individual import solution_specifications
```

```python
all_scores = set()
for score0, score1 in solution_specifications:
    all_scores = all_scores.union({score0, score1})
all_scores = sorted(list(all_scores))
```

```python
matrix = pd.DataFrame(index=all_scores, columns=all_scores)
```

```python
for score0, score1 in solution_specifications:
    matrix.loc[score0, score1] = 1
    matrix.loc[score1, score0] = 1
    matrix.loc[score0, score0] = 1
    matrix.loc[score1, score1] = 1
```

```python
matrix
```

```python
mask = pd.isna(matrix).stack()
mask = mask.loc[mask].reset_index(drop=False)
no_solution = set()
for idx, row in mask.iterrows():
    no_solution.add(tuple(sorted(list([row['level_0'], row['level_1']]))))
```
