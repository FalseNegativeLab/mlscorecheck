```python
from mlscorecheck.check import (check_multiple_datasets_som_kfold_som_scores)
from mlscorecheck.core import (IntervalUnion)
from mlscorecheck.utils import (calculate_scores,
                                generate_problems_with_folds)
```

```python
k = 4
eps = 10**(-k)
```

```python
foldings, problems = generate_problems_with_folds(n_problems=3,
                                                    n_repeats=2,
                                                    random_seed=5)
```

```python
foldings, problems
```

```python
scores = calculate_scores(foldings, 
                            strategy=('som', 'som'), 
                            rounding_decimals=k,
                            scores_only=True)
```

```python
scores
```

```python
flag, details = check_multiple_datasets_som_kfold_som_scores(scores=scores, 
                                                                eps=eps, 
                                                                datasets=problems,
                                                                return_details=True)
```

```python
assert flag
```

```python
for tmp in details['succeeded']:
    assert tmp['consistency']
    for item in tmp['details']:
        if item.get('target_interval_reconstructed') is not None and item.get('consistency'):
            target_interval = item['target_interval_reconstructed']
            iun = IntervalUnion(target_interval)
            assert not iun.intersection(IntervalUnion(item['target_interval'])).is_empty()
```

```python
scores['bacc'] = 0.9
```

```python
flag, details = check_multiple_datasets_som_kfold_som_scores(scores=scores, 
                                                                eps=eps, 
                                                                datasets=problems,
                                                                return_details=True)
```

```python
assert not flag
```
