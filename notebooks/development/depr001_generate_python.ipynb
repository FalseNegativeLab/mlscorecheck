```python
module_doc =\
'"""\n'\
'This module is generated by the notebook notebook/utils/001_generate_python.ipynb\n'\
'"""\n'\
'\n'\
'import numpy as np\n\n'\
'from ._scores import sqrt, reduce_solutions\n'
```

```python
# removing all the solutions containing sympy 'zoo' - complex infinity
for sol in solutions:
    sol.remove_zoo_solutions()

# filtering out the degenerate problems (having solutions tp=p tn=0 or variations of this)
# independent from the scores
solutions = [sol for sol in solutions if not sol.degenerate_problem()]
```

```python
names = []
for sol in solutions:
    names.extend(sol.names())
```

```python
all_func = ''
for name in names:
    all_func = all_func + f"  '{name}',\n"
```

```python
all_func =\
f'__all__ = [{all_func}]\n'
```

```python
with open('_solutions.py', 'w') as handle:
    handle.write(module_doc)
    handle.write('\n')
    handle.write(all_func)
    handle.write('\n')
    for sol in solutions:
        handle.write(sol.to_python())
```
