```python
from mlscorecheck.check import (check_scores)
from mlscorecheck.core import (IntervalUnion)
from mlscorecheck.utils import (generate_problems_with_folds,
                                calculate_scores)
```

```python
k = 4
eps = 10**(-k)

folds, problem = generate_problems_with_folds(n_folds=1, n_repeats=1, random_seed=5)
folds, problem
```

```python
scores = calculate_scores(folds, 
                            rounding_decimals=k)
scores
```

```python
flag, details = check_scores(scores, 
                                eps=eps, 
                                dataset=problem,
                                return_details=True)
```

```python
assert flag
```

```python
for tmp in details['succeeded']:
    assert tmp['consistency']
    for item in tmp['details']:
        if item.get('target_interval_reconstructed') is not None and item.get('consistency'):
            target_interval = item['target_interval_reconstructed']
            iun = IntervalUnion(target_interval)
            assert not iun.intersection(IntervalUnion(item['target_interval'])).is_empty()
```

```python
scores['bacc'] = 0.9
```

```python
flag, details = check_scores(scores, 
                                eps=eps, 
                                dataset=problem, 
                                return_details=True)
```

```python
assert not flag
```
