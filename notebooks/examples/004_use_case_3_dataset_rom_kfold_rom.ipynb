```python
import numpy as np

from mlscorecheck.check import (check_multiple_datasets_rom_kfold_rom_scores)

from mlscorecheck.core import (score_functions_with_solutions,
                                load_solutions,
                                Interval,
                                IntervalUnion)

from mlscorecheck.utils import (generate_problem,
                                generate_n_problems,
                                calculate_scores_rom,
                                generate_folding_problem,
                                calculate_all_scores)
```

```python
k = 4
eps = 10**(-k)
```

```python
foldings, all_foldings, problems = [], [], []

for _ in range(5):
    folding, problem = generate_folding_problem(n_repeats=(np.random.randint(20)%5+1))
    foldings.append(folding)
    all_foldings.extend(folding)
    problems.append(problem)
```

```python
scores = calculate_scores_rom(all_foldings, rounding_decimals=k)
```

```python
flag, details = check_multiple_datasets_rom_kfold_rom_scores(scores=scores, 
                                                                eps=eps, 
                                                                setup=problems,
                                                                return_details=True)
```

```python
flag
```

```python
for tmp in details['succeeded']:
    assert tmp['consistency']
    for item in tmp['details']:
        if item.get('target_interval_reconstructed') is not None and item.get('consistency'):
            target_interval = item['target_interval_reconstructed']
            iun = IntervalUnion(target_interval)
            assert not iun.intersection(IntervalUnion(item['target_interval'])).is_empty()
```
